```python
import matplotlib.pyplot as plt

# Data
variety_counts = {
    'Pinot Noir': 146,
    'Zinfandel': 134,
    'Cabernet Sauvignon': 133,
    'Sauvignon Blanc': 125,
    'Riesling': 119,
    'Chardonnay': 117,
    'Merlot': 113,
    'Syrah': 113
}

# Extract labels and values
varieties = list(variety_counts.keys())
counts = list(variety_counts.values())

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c2f0c2', '#ff6666']
# Bar chart
plt.figure(figsize=(10, 6))
plt.barh(varieties, counts, color=colors)
plt.xlabel('Count')
plt.ylabel('Variety')
plt.title('Distribution of Grape Varieties')
plt.gca().invert_yaxis()  # Invert y-axis to show the largest count on top
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Verify if 'variety' column exists
if 'variety' in data.columns:
    # Count values for the 'variety' column
    variety_counts = data['variety'].value_counts().reset_index()
    variety_counts.columns = ['variety', 'count']

    # Define custom colors
    colors = ['violet', 'mediumseagreen', 'ivory', 'chocolate',
              'mintcream', 'lightsalmon', 'purple', 'aqua']

    # Create the bar plot
    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=variety_counts,
        x='variety',
        y='count',
        palette=colors,
        hue='variety',
        dodge=False
    )

    # Remove legend because we're not using `hue` for grouping
    plt.legend([], [], frameon=False)

    # Customize the plot
    plt.title('Count of Grape Variety', fontsize=16, color='yellowgreen', loc='center')
    plt.xlabel('Grape Variety', fontsize=14, color='peru')
    plt.ylabel('Count', fontsize=14, color='peru')
    plt.xticks(rotation=45, fontsize=12)
    plt.yticks(fontsize=12)

    # Add counts on top of bars
    for i, count in enumerate(variety_counts['count']):
        plt.text(i, count + 1, str(count), ha='center', va='bottom', fontsize=12, color='teal')

    # Show the plot
    plt.tight_layout()
    plt.show()
else:
    print("Error: 'variety' column not found in the data.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check if required columns exist in the DataFrame
if all(col in data.columns for col in ['region', 'quality_score', 'quality_category']):
    # Create a grouped bar plot
    plt.figure(figsize=(12, 8))
    sns.barplot(
        data=data, 
        x='region', 
        y='quality_score', 
        hue='quality_category',  # Groups bars by quality category
        palette='viridis'
    )

    # Customize the plot
    plt.title('Regionwise Harvest Of Various Quality Of Grapes', fontsize=16, color='green', loc='center')
    plt.xlabel('Region', fontsize=14)
    plt.ylabel('Quality Score', fontsize=14)
    plt.legend(title='Quality Category', title_fontsize=12, fontsize=10)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    # Show the plot
    plt.show()
else:
    print("Error: Ensure the DataFrame contains 'region', 'quality_score', and 'quality_category' columns.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
correlation = data.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(
    correlation, 
    annot=True,       # Show correlation values
    cmap='coolwarm',  # Color map
    fmt='.2f',        # Format for values
    linewidths=0.5,   # Line width between cells
    cbar=True         # Show color bar
)

# Add title
plt.title('Correlation Matrix Heatmap', fontsize=16)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Assuming 'data' is your DataFrame
# Create a dictionary for the mapping
category_map = {
    "Low": 0,
    "Medium": 1,
    "High": 2,
    "Premium": 3
}

# Apply the mapping to the 'quality_category' column
data['quality_category_numeric'] = data['quality_category'].map(category_map)

# Calculate the correlation between 'quality_score' and the numeric 'quality_category'
correlation = data['quality_score'].corr(data['quality_category_numeric'])
print(correlation)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Create the correlation matrix for the relevant numerical columns
corr_matrix = data[['quality_score', 'quality_category_numeric', 'sugar_content_brix', 
                     'acidity_ph', 'cluster_weight_g', 'berry_size_mm', 'sun_exposure_hours', 
                     'soil_moisture_percent', 'rainfall_mm']].corr()

# Set up the plot
plt.figure(figsize=(10, 8))

# Create a heatmap from the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)

# Title for the heatmap
plt.title('Correlation Matrix Heatmap')

# Display the heatmap
plt.show()
```
